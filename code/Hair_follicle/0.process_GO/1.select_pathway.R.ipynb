This script describes how we extracted descendent GO terms of "Hair follicle development (GO:0001942)" and their related genes.

```r
library(dplyr)
setwd('/nfs/public/xixi/scRegulate/SHARE-seq')
```

# Filter GO and GOA according to the single cell dataset

```r
rna <- read.csv('skin.rna.aggregate_30cells.csv', row.names = 1)
rna
```

```r
go <- read.csv("/nfs/public/xixi/scRegulate/go/go-organize.txt", header = TRUE, stringsAsFactors = FALSE, sep = '\t')
go_annotation <- read.csv("/nfs/public/xixi/scRegulate/go/go_annotation.csv", stringsAsFactors = FALSE)
```

```r
gos <- unique(go_annotation$X4)
print(length(gos))
gos_keep = c()
for (i in 1:length(gos)){
    temp <- go_annotation[go_annotation$X4==gos[i],]
    if (length(intersect(temp$X2, row.names(rna)))>1){
        gos_keep = c(gos_keep, gos[i])
    }
}
print(length(gos_keep))
```

```r
go_annotation_filtered <- go_annotation[go_annotation$X4 %in% gos_keep,]
go_annotation_filtered
```

```r
go_filtered <- go[go$id1 %in% unique(go_annotation_filtered$X4),]
go_filtered <- go_filtered[go_filtered$id2 %in% unique(go_annotation_filtered$X4),]
go_filtered
```

```r
write.table(go_filtered, '/nfs/public/xixi/scRegulate/go/go_filtered_shareseq.txt', sep = '\t', quote=FALSE, row.names = FALSE)
```

```r
write.table(go_annotation_filtered, '/nfs/public/xixi/scRegulate/go/goa_filtered_shareseq.txt', sep = '\t', quote=FALSE, row.names = FALSE, 
           col.names=F)
```

# Extract GO terms and genes related to "Hair follicle development"

```r
go_filtered <- read.csv('/nfs/public/xixi/scRegulate/go/go_filtered_shareseq.txt', sep = '\t')
go_filtered
```

```r
go_annotation_filtered <- read.csv('/nfs/public/xixi/scRegulate/go/goa_filtered_shareseq.txt', sep = '\t', header = F)
go_annotation_filtered
```

```r
genes <- row.names(rna)
```

```r
go_sub <- go_filtered[go_filtered$id2=="GO:0001942",]
count <- 0
rowsum <- 0
while (1){
    temp <- go_filtered[go_filtered$id2 %in% go_sub$id1,]
    if (sum(!duplicated(rbind(go_sub, temp)))>rowsum){
        go_sub <- rbind(go_sub, temp)
        go_sub <- go_sub[!duplicated(go_sub),]
        rowsum <- nrow(go_sub)
    }else{
        break
    }
   count = count + 1 
}
go_sub
```

```r
length(unique(c(go_sub$id1, go_sub$id2)))
```

```r
go_annotation_filtered_sub <- go_annotation_filtered[go_annotation_filtered$X8=='P',]
go_annotation_filtered_sub <- go_annotation_filtered_sub[go_annotation_filtered_sub$X4 %in% unique(c(go_sub$id1, go_sub$id2)),]
sum(unique(go_annotation_filtered_sub$X2) %in% genes)
```

```r
go_annotation_filtered_sub <- go_annotation_filtered_sub[go_annotation_filtered_sub$X2 %in% genes,]
go_annotation_filtered_sub
```

```r
write.table(go_sub, "/nfs/public/xixi/scRegulate/SHARE-seq/go/go_follicle.txt", quote = FALSE, row.names = FALSE, sep = '\t')
write.table(go_annotation_filtered_sub, "/nfs/public/xixi/scRegulate/SHARE-seq/go/goa_follicle.txt", quote = FALSE, sep = '\t', row.names = F)
```
